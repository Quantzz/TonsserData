# Filter for Arsenal

```python
#Fitler dataframe for all matches were Arsenal plays.
afc_df = raw_df[(raw_df['HomeTeam'] == "Arsenal") | (raw_df['AwayTeam'] == "Arsenal")]
afc_df.index = pandas.to_datetime(afc_df.index, format='%d/%m/%y')

print(afc_df.shape)
afc_df.head()
```

# Ploting of Arsenal goals

```python
# Helperfunction to determine if arsenal is home or away.
def arsenal_goals(row):
    if row['HomeTeam'] == 'Arsenal':return row['FTHG']
    if row['AwayTeam'] == 'Arsenal':return row['FTAG']
    else: return None
    
afc_goals = afc_df.apply(lambda row: arsenal_goals(row), axis=1);
```

```python
fig, ax = plt.subplots(figsize=(15,7))
afc_goals.plot(kind = 'bar', ax=ax, color = 'blue', title="Goals per game for Arsenal")
ax.set_ylabel("Goals");
```

```python
month_goals = afc_goals.groupby(by=[afc_goals.index.month]).sum()
```

```python
fig, ax = plt.subplots(figsize=(15,7))
month_goals.plot(kind = 'bar', ax=ax, color = 'blue', title="Goals per game for Arsenal for a given month")
ax.set_ylabel("Goals");
ax.set_xlabel("Month #");
```
